```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
d = pd.read_csv('HTRU_2.csv', header=None)
```

Pulsars are a type of Neutron star that produce a detectable, periodically repeatable, pattern of broadband radio emission as they rotate. Using large radio telescopes, we look for periodic radio signals to search for pulsars. 

# Problem 1

The HTRU_2 dataset describes 8 continuous variables of a sample of pulsar candidates. 4 of the 8 continuous variables are examined here:
1.  Mean of the DM-SNR curve.
2.  Standard deviation of the DM-SNR curve.
3.  Excess kurtosis of the DM-SNR curve.
4.  Skewness of the DM-SNR curve.
In addition to the 8 continuous variables, the candidates of the dataset are described by 1 single class variable, where legitimate pulsar examples are in the positive class and the false pulsar examples are in the negative class. We may than compare each variable betweeen these two classes. 

#### Mean of the DM-SNR curve

```python
mean_DMSNR_0 = d.loc[d[8] == 0, 4]
mean_DMSNR_1 = d.loc[d[8] == 1, 4]

mean0 = mean_DMSNR_0.mean()
std0 = mean_DMSNR_0.std()
mean1 = mean_DMSNR_1.mean()
std1 = mean_DMSNR_1.std()

fig, ax = plt.subplots(1, 1, figsize=(8,6))

plt.xlabel('DM-SNR Curve Mean')
plt.ylabel('Density')
plt.title('Normalized Histograms of the Candidates Means of the DM-SNR Curve')

ax.hist(mean_DMSNR_0, bins=30, density=True, alpha=0.5)
ax.hist(mean_DMSNR_1, bins=30, density=True, alpha=0.5)

ax.legend(['Class 0  \nmean=' + str(mean0) + '\nstandard deviation=' + str(std0),
           '\nClass 1  \nmean=' + str(mean1) + '\nstandard deviation=' + str(std1)])
```

We can see by the histogram that the average of the class 0 observations' DM-SNR curve means is significantly less than that of the class 1 observations' DM-SNR curve means. Additionally, the distribution of the class 0 histogram has less spread than class 1's, with standard deviations of 24.4 and 45.2, respectively. 

#### Standard Devation of the DM-SNR Curve

```python
sd_DMSNR_0 = d.loc[d[8] == 0, 5]
sd_DMSNR_1 = d.loc[d[8] == 1, 5]

mean0 = sd_DMSNR_0.mean()
std0 = sd_DMSNR_0.std()
mean1 = sd_DMSNR_1.mean()
std1 = sd_DMSNR_1.std()

fig, ax = plt.subplots(1, 1, figsize=(8,6))

plt.xlabel('DM-SNR Curve Standard Deviation')
plt.ylabel('Density')
plt.title('Normalized Histograms of the Candidates Standard Deviations of the DM-SNR Curve', wrap=True)


ax.hist(sd_DMSNR_0, bins=30, density=True, alpha=0.5)
ax.hist(sd_DMSNR_1, bins=30, density=True, alpha=0.5)

ax.legend(['Class 0  \nmean=' + str(mean0) + '\nstandard deviation=' + str(std0),
           '\nClass 1  \nmean=' + str(mean1) + '\nstandard deviation=' + str(std1)])
```

We can see from the histograms that the mean DM-SNR curve standard deviation is much lower in the class 0 histogram than the class 1 histogram. 

#### Excess Kurtosis of the DM-SNR Curve

```python
ex_DMSNR_0 = d.loc[d[8] == 0, 6]
ex_DMSNR_1 = d.loc[d[8] == 1, 6]

mean0 = ex_DMSNR_0.mean()
std0 = ex_DMSNR_0.std()
mean1 = ex_DMSNR_1.mean()
std1 = ex_DMSNR_1.std()

fig, ax = plt.subplots(1, 1, figsize=(8,6))

plt.xlabel('DM-SNR Curve Excess Kurtosis')
plt.ylabel('Density')
plt.title('Normalized Histograms of the Candidates Excess Kurtosis of the DM-SNR Curve')

ax.hist(ex_DMSNR_0, bins=30, density=True, alpha=0.5)
ax.hist(ex_DMSNR_1, bins=30, density=True, alpha=0.5)

ax.legend(['Class 0  \nmean=' + str(mean0) + '\nstandard deviation=' + str(std0),
           '\nClass 1  \nmean=' + str(mean1) + '\nstandard deviation=' + str(std1)])
```

We can see that the mean DM-SNR Curve Exesss Kurtosis for the class 1 observations is smaller than that of the class 0 observations. 

#### Skew of the DM-SNR Curve

```python
skew_DMSNR_0 = d.loc[d[8] == 0, 7]
skew_DMSNR_1 = d.loc[d[8] == 1, 7]

mean0 = skew_DMSNR_0.mean()
std0 = skew_DMSNR_0.std()
mean1 = skew_DMSNR_1.mean()
std1 = skew_DMSNR_1.std()

fig, ax = plt.subplots(1, 1, figsize=(8,6))

plt.xlabel('DM-SNR Curve Skew')
plt.ylabel('Density')
plt.title('Normalized Histograms of the Candidates Skews of the DM-SNR Curve')


ax.hist(skew_DMSNR_0, bins=30, density=True, alpha=0.5)
ax.hist(skew_DMSNR_1, bins=30, density=True, alpha=0.5)

ax.legend(['Class 0  \nmean=' + str(mean0) + '\nstandard deviation=' + str(std0),
           '\nClass 1  \nmean=' + str(mean1) + '\nstandard deviation=' + str(std1)])
```

The mean DM-SNR curve skew is smaller for class 1 observations than class 0 observations. 

# Problem 2

```python
pos = d.loc[d[8] == 1]
PA = len(pos) / len(d[8])
print('The probability of an observation being a pulsar candidate is P(A)=', PA)
```

```python
mean80 = d.loc[d[0] <= 80]
PB = len(mean80) / len(d[0])
print('The probability that an observation with Mean of the integrated profile <80 is P(B)=', PB)
```

```python
PAnB = PA * PB
print('The probability that an observation satisfies both of the conditions above, A and B, is P(AnB)=', PAnB)
```

# Problem 3

```python
P_A_given_B = PAnB / PB
print('For an observation with mean of the integrated profile < 80, the probability that it is a pulsar candidate is
      P(A|B)=', P_A_given_B)
```

This is the same value as the probability of an observation being a pulsar candidate, P(A). <br>
Thus, <br>
$P(A)=P(A|B)$

```python
P_B_given_A = PAnB / PA
print('For a pulsar candidate, the probability that its Mean of the integrated profile < 80 is P(B|A)=', P_B_given_A)
```

This is the same value as the probability that an observation with mean of the integrated profile <80, P(B). <br>
Thus, <br>
$ P(A \cap B) = P(B)$
