```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
d = pd.read_csv('HTRU_2.csv', header=None)
print(d)
x = d.loc[d[8]==0,0]
print(x)
```

# Problem 1

```python
mean_DMSNR_0 = d.loc[d[8]==0, 4]
mean_DMSNR_1 = d.loc[d[8]==1, 4]

mean0 = str(mean_DMSNR_0.mean())
std0 = str(mean_DMSNR_0.std())
mean1 = str(mean_DMSNR_1.mean())
std1 = str(mean_DMSNR_1.std())

fig, ax = plt.subplots(1,1)
ax.hist(mean_DMSNR_0, bins=30, density=True, alpha=0.5)
ax.hist(mean_DMSNR_1, bins=30, density=True, alpha=0.5)
plt.xlabel('DM-SNR Curve Mean')
plt.ylabel('Density')
ax.legend(['Class 0' + ' mean=' + mean0 + ' std=' + std0, 'Class 1' + ' mean=' + mean1 + ' std=' + std1], fontsize=8)
```

```python
pos = d.loc[d[8]==1]
PA = len(pos)/len(d[8])
print('The probability of an observation being a pular candidate is P(A)=', PA)
```

```python
mean80=d.loc[d[0]<=80]
PB = len(mean80)/len(d[0])
print('The probability that an observation with Mean of the integrated profile <80 is P(B)=', PB)
```

```python
PAnB = PA * PB
print('The probability that an observation satisfies both of the conditions above, A and B, is P(AnB)=', PAnB)
```

```python
P_A_given_B = PAnB / PB
print('For an obersvation with mean of the integrated profile < 80, the probability that it is a pulsar candidate is', P_A_given_B)
```
