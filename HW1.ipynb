```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

R = norm.rvs(loc = 3.0, scale = 0.01, size = 100000)
fig, ax = plt.subplots(1, 1)
ax.hist(R, 50, density=True)
x = np.linspace(2.95, 3.05, 1000)
ax.plot(x, norm.pdf(x, loc=3.0, scale=0.01), linewidth=5, alpha=0.6)
plt.xlim([2.95, 3.05])
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
ax.hist(R, 50, density=True)
plt.yscale('log')
x = np.linspace(2.95, 3.05, 1000)
ax.plot(x, norm.pdf(x, loc=3.0, scale=0.01), linewidth=5, alpha=0.6)
plt.xlim([2.95, 3.05])
plt.show()
```
